```python
from sklearn import cross_validation,datasets,grid_search,linear_model,metrics
import numpy as np
import pandas as pd
```

```python
iris=datasets.load_iris()
```

```python
train_data,test_data,train_labels,test_labels=cross_validation.train_test_split(iris.data,iris.target,test_size=0.3,random_state=0)
```

```python
classifier=linear_model.SGDClassifier(random_state=0)
```

```python
classifier.get_params().keys()
```

```python
parameters_grid={"loss":["hinge","log","squared_hinge","squared_loss"],"penalty":["l1","l2"],"n_iter":range(5,10),
                 "alpha":np.linspace(0.0001,0.01,num=5)}
```

```python
cv=cross_validation.StratifiedShuffleSplit(train_labels,n_iter=10,test_size=0.2,random_state=0)
```

```python
grid_cv=grid_search.GridSearchCV(classifier,parameters_grid,scoring="accuracy",cv=cv)
```

```python
%%time
grid_cv.fit(train_data,train_labels)
```

```python
grid_cv.best_estimator_
```

```python
print grid_cv.best_score_
print grid_cv.best_params_
```

```python
grid_cv.grid_scores_[:10]
```

```python
randomized_grid_cv=grid_search.RandomizedSearchCV(classifier,parameters_grid,scoring="accuracy",cv=cv,n_iter=20,random_state=0)
```

```python
%%time
randomized_grid_cv.fit(train_data,train_labels)
```

```python
print randomized_grid_cv.best_score_
print randomized_grid_cv.best_params_
```
