GПодключение необходимых модулей

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

Генерация 1000  значений из экспоненциального распределения

```python
lamda=5.0
expon_rv=sts.expon(scale=1.0/lamda)
values=expon_rv.rvs(1000)
```

Построение гистограммы експоненциального распределения с выводом теоретической плотности распределения вероятности 

```python
plt.hist(values,bins=20,normed=True)
x=np.linspace(0,1.4,100)
pdf=expon_rv.pdf(x)
plt.plot(x,pdf,label="theoretical PDF",alpha=1)
plt.legend()
plt.ylabel("$f(x)$")
plt.xlabel("$x$")
```

Генерация 1000 выборок размера 5 и вычисление для них выборочных средних

```python
means=[]
for i in range(1000):
    rv=sts.expon(scale=1.0/lamda)
    rv_values=rv.rvs(5)
    rv_mean=np.mean(rv_values)
    means.append(rv_mean)
arrmeans=np.array(means)
```

Построение гистограммы распределения выборочных средних и теоритической плотности распределения для приближающего нормального распределения

```python
plt.hist(arrmeans,bins=20,normed=True)
rv_norm=sts.norm(loc=1.0/lamda,scale=np.sqrt((1.0/lamda**2)/5.0))
x=np.linspace(0,0.4,100)
pdf=rv_norm.pdf(x)
plt.plot(x,pdf,label="theoretical norm PDF",alpha=1)
plt.legend()
plt.ylabel("$f(x)$")
plt.xlabel("$x$")
```
